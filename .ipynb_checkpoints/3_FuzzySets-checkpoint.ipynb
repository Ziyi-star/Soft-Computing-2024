# Fuzzy Sets

## Fuzzy-Mengen

1. Wählen Sie geeignete Zugehörigkeitsfunktionen aus, um die folgenden Fuzzy-Mengen abzubilden:
  * Kalt, Kühl, Warm, Heiß

```python
# the softcomputing.fuzzy module contains a selection of membership functions
from softcomputing.fuzzy import *
import numpy as np
import matplotlib.pyplot as plt

xs = np.arange(-10,40,0.1)
mu_kalt = z_shape(-5,0)
mu_kalt(xs)

mu_kuhl = triangular(0,8,16)
mu_kuhl(xs)

mu_warm = triangular(8,16,28)
mu_warm(xs)

mu_heiss = sigmoid(30,1)
#m Wendepunkt, a Strauchung
mu_heiss(xs)

# Bilder zeichnen
plt.figure(figsize=(8,2))
plt.plot(xs,mu_kalt(xs),label="$\mu_{kalt}$")
plt.plot(xs,mu_kuhl(xs),label="$\mu_{kühl}$")
plt.plot(xs,mu_warm(xs),label="$\mu_{warm}$")
plt.plot(xs,mu_heiss(xs),label="$\mu_{heiss}$")
plt.legend()#label anzeigen in Legende
plt.tight_layout()
plt.show()
```

2. Implementieren Sie *linguistische Modifizierer* für ihre Mengen, z.B.
  * `sehr`, oder `weniger`

```python
import math
xs = np.arange(-10,40,0.1)


#folien 35
def sehr(mu):
    return lambda x:mu(x)**2

def weniger(mu):
    #return lambda x: math.sqrt(mu(x))
    return lambda x:np.sqrt(mu(x))

mu_sehr_kuhl = sehr(mu_kuhl)
mu_weniger_kuhl = weniger(mu_kuhl)

# Bilder zeichnen
plt.figure(figsize=(8,2))
plt.plot(xs,mu_sehr_kuhl(xs),label="$\mu_{sehr_kuhl}$")
plt.plot(xs,mu_weniger_kuhl(xs),label="$\mu_{weniger_kuhl}$")
plt.plot(xs,mu_kuhl(xs),label="$\mu_{kuhl}$")
plt.legend()#label anzeigen in Legende
plt.tight_layout()
plt.show()
```

3. Geben Sie jeweils Support und Toleranz ihrer Mengen an.
  * Sie können sich an der Definition für eine Implementierung orientieren, z.B. mittels einer List-Comprehension.

```python
xs = np.arange(-10,40,0.1)
support_kuhl = np.array([x for x in xs if mu_kuhl(x) > 0])
tolenz_kuhl = np.array([x for x in xs if mu_kuhl(x) >= 0.98])

# Bilder zeichnen
plt.figure(figsize=(8,2))
plt.plot(xs,mu_kuhl(xs),label="$\mu_{kuhl}$")
#scatter punkt zeichnen
plt.scatter(support_kuhl, mu_kuhl(support_kuhl),color = 'yellow')
plt.scatter(tolenz_kuhl, mu_kuhl(tolenz_kuhl),color = 'red')
plt.legend()#label anzeigen in Legende
plt.tight_layout()
plt.show()
```
